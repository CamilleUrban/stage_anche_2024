```python
import numpy as np
import librosa
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dossier = 'MAD Dynamique/mes_2024.03.20/'
x = pd.read_csv(dossier + 'Anche1', delimiter='\t')
```

```python
plt.plot(x['Temps'], x['Canal D'])
```

```python
y = np.array(x['Canal D'])
sr= 0.0001
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,fmax=8000)
```

```python
print(y.shape)
CGS = librosa.feature.spectral_centroid(y=y, sr=sr)
```

```python
plt.plot(CGS.squeeze())
#print(CGS.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#X données
X = np.array(x['Canal A'])

reg = LinearRegression()
reg.fit(X,CGS)
reg.score(X,CGS)
```
